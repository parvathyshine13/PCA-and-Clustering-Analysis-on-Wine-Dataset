#### Dimensionality Reduction with PCA:

```python
# Standardizing the Features :

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_scaled = sc.fit_transform(df[cols])
```

```python
# Fiting PCA on scaled data

from sklearn.decomposition import PCA
pca= PCA()
pca.fit(df_scaled)
```

```python
# Screeplot to determine optimal number of components (hom much variance in each PC)

# Explained variance ratio ( proportion of total variance by each principal component tv+tv=1) 
#                          Full of arrays
ex_v = pca.explained_variance_ratio_

#Plotting the Screeplot:

plt.plot(range(1, len(ex_v) + 1), pca.explained_variance_ratio_, marker='o')
#        range(1, len(...) + 1) = starting from 1 to PCs

plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
# Cumulative Explained Variance [variance*(pc1,pc1+pc2,pc1+pc2+pc3....)]


cu_v = np.cumsum(ex_v)  # sum variance progressively

plt.plot(range(1, len(cu_v) + 1), cu_v, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% Explained Variance') #To see the point reaches the 95% Threshold
plt.legend()
plt.grid(True)
plt.show()

# Find the index where cumulative variance >= 0.95
hit_at_95 = np.argmax(cu_v >= 0.95) + 1 # argmax - find the first principal component that crosses the 95% variance threshold.
print(f"Number of components to reach 95% variance: {hit_at_95}")
```

```python
#Transform scaled data to Principal Components.

pca = PCA(n_components= 10)  # From the figure
df_pca = pca.fit_transform(df_scaled)
print("Transformed dataset shape:", df_pca.shape)
```

### Clustering with Original Data:

```python
#Clustering with Original Plot

# Applying KMeans (OG):
from sklearn.cluster import KMeans
kmeans_og = KMeans(n_clusters=3, random_state=42) 
kmeans_og.fit(df_scaled)
labels_og = kmeans.labels_ # Add the cluster labels to the data

# Visualising the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels_og, cmap='plasma', s=50)
plt.title("K-Means Clustering (Original Data)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.show()
```

```python
#Silhouette Score: Measures how well each point fits in its cluster (higher is better, range: -1 to 1)

from sklearn.metrics import silhouette_score
s_score = silhouette_score(df_scaled, labels_og)
print(f"Silhouette Score: {s_score:.3f}")

# Daviesâ€“Bouldin Index: Lower score means better clustering

from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(df_scaled, labels_og)
print(f"Daviesâ€“Bouldin Index: {db_index:.3f}")
```

#### Clustering with PCA Data:


```python
#Clustering with PCA Data:

# Applying KMeans (PCA):
from sklearn.cluster import KMeans
kmeans_pca = KMeans(n_clusters=3, random_state=42) 
kmeans_pca.fit(df_pca)
labels_pca = kmeans.labels_ # Add the cluster labels to the data

# Visualising the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels_pca, cmap='plasma', s=50)
plt.title("K-Means Clustering (PCA data)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.show()
```

```python
#Silhouette Score:       Higher Silhouette Score= Clusters are more distinct and well-separated.

from sklearn.metrics import silhouette_score
s_score_pca = silhouette_score(df_pca, labels_pca)
print(f"Silhouette Score: {s_score_pca:.3f}")

# Daviesâ€“Bouldin Index:   Lower Daviesâ€“Bouldin Index = Clusters are less similar to each other.

from sklearn.metrics import davies_bouldin_score
db_index_pca = davies_bouldin_score(df_pca, labels_pca)
print(f"Daviesâ€“Bouldin Index: {db_index_pca:.3f}")
```

Inference from comparing both :

<br>Silhouette Score (Higher is Better) - Ranges from -1 to 1
<br>Daviesâ€“Bouldin Index (Lower is Better) 

<br>Comparing both the scores - PCA has better data for clustering : Clusters are more distinct and less similar to each other.

### Comparison and Analysis:

##### Similarities:

1. Overall cluster approach and number of clusters remained the same.
2. Cluster structures are formed in both.

##### Differnces:

1. The cluster quality improved with PCA due to less noise and redundant features , cleaner separation btw clusters in fewer dimensions.
2. Visualisations after PCA are easier to interpret.

##### Impact of Dimensionality Reduction (PCA):

1. Improved clustering performance and Simplified visualization
2. Noise reduction: PCA removes irrelevant variance, focusing on what really matters in the data.
3. Faster computation: Fewer dimensions mean quicker clustering.
4. It loses direct Interpretablity of features.

##### Trade -Offs - PCA vs Original Data

1. Accuracy : PCA is better due to noise reduction.
2. Interpretablity : PCA is better to visualise.
3. Feature Meaning : Original data are interpretable but PCA components are abstract.
4. Performance : Faster clustering with PCA and reduces computational cost too.
5. Overfitting Risk : PCA has reduced overfitting.

###### If every feature matters, use original data . for better cluster quality and performance, use PCA.
